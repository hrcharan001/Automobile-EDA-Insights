## Data cleaning

```python
# lets check if there are any null values in the columns
df.isnull().sum()
```

```python
# lets check duplicates if available
df.duplicated().sum()  # there are no duplicates
```

```python
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
```

```python
pd.reset_option('display.max_rows',None)
pd.reset_option('display.max_columns',None)
```

```python
df
```

```python
# lets replace some of the row values which is missing
#fill numerical missing values
numerical_columns = df.select_dtypes(include=np.number).columns

for col in numerical_columns:
    df[col].fillna(df[col].mean(), inplace=True)
```

```python
df
```

```python
# lets fill categorical missing values
categorical_columns = df.select_dtypes(include='object').columns

for col in categorical_columns:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

## Converting datatypes 

```python
# change price column ? from to NAN
df['price'] = df['price'].replace('?',np.nan)
df
```

```python
# convert price column datatype to numeric
df['price'] = pd.to_numeric(df['price'], errors='coerce') #errors=coerce converts invalid values into NAN
```

```python
# filling missing values
df['price'].fillna(df['price'].mean(), inplace=True)
```

```python
df
```

```python
print(df['price'].dtype)
```

```python
# TO convert all the other required column to numerical datatype using loops
col = ['normalized-losses','bore','stroke','horsepower','peak-rpm']

for columns in col:
         # change price column ? from to NAN
    df[columns] = df[columns].replace('?',np.nan)

    # converting the column to numeric
    df[columns] = pd.to_numeric(df[columns],errors='coerce')

    # now will fill the NAN values with the mean
    df[columns].fillna(df[columns].mean(),inplace=True)


     
```

## Data Visualization 

### Histogram (univariate analysis)

```python
sns.histplot(x = df['price'],kde= True, bins=10,color='black',edgecolor='red')
plt.title("Price Variation")
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.figure(figsize=(10,6))
plt.show()
```

## Company wise Sales 

```python
sales.plot(kind='barh', color='black',edgecolor='red')

plt.title("Company Wise Sales")

plt.xlabel('Sales')

plt.ylabel('Make')

plt.yticks(rotation=20)

plt.show()
```

##  Scatter Plot

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='horsepower',y='city-mpg',data=df,hue='make')
plt.title('Horsepower vs City-MPG')
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.lineplot(data=df,x="length",y ="engine-size",ci=None)

plt.xlabel('Length')

plt.ylabel('Engine Size')

plt.title('Length vs Engine size')
```
